```python
import csv

values = []
def to_signed(val, bits):
    if val >= 2**(bits-1):
        return val - 2**bits
    else:
        return val

with open('run-tb_dds_core-default/waveform.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        values.append(to_signed(int(row[0]), 16))
        # values.append(int(row[0]))

# plot the values
import matplotlib.pyplot as plt
plt.plot(values)
plt.show()
```
